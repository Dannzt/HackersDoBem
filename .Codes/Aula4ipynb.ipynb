```python
#exemplo 2 Media idade
id1 = input("Digite a primeira idade: \n")
id2 = input("Digite a segunda idade: \n")
id3 = input("Digite a terceira idade: \n")
id4 = input("Digite a quarta idade: \n")
id5 = input("Digite a quinta idade: \n")

id1 = int(id1)
id2 = int(id2)
id3 = int(id3)
id4 = int(id4)
id5 = int(id5)

media = (id1 + id2 + id3 + id4 + id5) / 5

print(f"A média das idades é: {media}")
```

**A mesma logica dos codigos acima mas usando while**

```python
#Exemplo 1 tabuada
numero = input("Digite um número para o cálculo de sua tabuada: \n")
numero = int(numero)

contador = 1
while (contador <=10):
  print(f"{numero} x {contador} = {numero * contador}")
  contador = contador + 1
```

```python
#Exemplo 2 media idade
contador = 1
soma = 0
while (contador <=5):
  idade = int (input(f"Digite a idade do °{contador} usuario \n"))
  soma = soma + idade
  contador = contador + 1

print(f"A média das idades é: {soma / 5}")
```

```python
# range (inicio,final, incremento)
#numeros impares
for i in range(1,11,2):
  print(i)
```

```python
# range (inicio,final, incremento)
#numeros pares
for i in range(0,11,2):
  print(i)
```

```python
#numeros impares ao contrario
for i in range(9,-1,-2):
  print(i)
```

```python
#numeros pares ao contrario
for i in range(10,-1,-2):
  print(i)
```
